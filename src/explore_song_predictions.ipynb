```python
import plotly.express as px 
import pandas
import song_predictor_1
```

```python
training_file_address = '../out/merged_data.csv'
model = song_predictor_1.get_song_predictor(file_address=training_file_address)
```

```python
address = "../out/clean_data.csv"
df = pandas.read_csv(address, encoding="MacRoman")
X = df.loc[
    :,
    [
        'release_year', 'duration_ms.x', 'explicit', 
        'mode', 'speechiness', 'acousticness',
        'instrumentalness', 'liveness', 'valence',
        'tempo', 'time_signature'
    ]
] 

predictions = model.predict(X)
```

```python
x_col_name = 'delta: sad to happy'
y_col_name = 'delta: calm to energetic'
z_col_name = 'delta: relaxed to stressed'
X[x_col_name] = predictions.transpose()[0]
X[y_col_name] = predictions.transpose()[1]
X[z_col_name] = predictions.transpose()[2]
X['name'] = df['name']
X.columns
```

```python
fig = px.scatter_3d(
    X, x = x_col_name, y = y_col_name, z = z_col_name,
    color = 'valence',
    hover_name='name', title='Delta Emotion Predictions', 
)

fig.show()
```

```python
df_survey = pandas.read_csv('../out/merged_data.csv')
```

```python
fig = px.scatter_3d(
    df_survey, x = x_col_name, y = y_col_name, z = z_col_name,
    color = 'valence', hover_name='name', title='Delta Emotion Survey Results', 
)

fig.show()
```

```python
# Analyzing the bias towards christmas songs.
christmas_indicators = ["christmas", "xmas", "winter", "jingle", "bells", "jolly", "little drummer boy"] 
count = 0
for name in df['name']:
    for indicator in christmas_indicators:
        if indicator in name.lower():
            print(name)
            count += 1
            break

print(count/len(X))
```
