This is the file where models are created and tested as pre-product research.

```python
import pandas
```

```python
df = pandas.read_csv('../out/merged_data.csv')
# df.head(3)
```

```python
labels = df.columns
# labels
```

```python
X = df.loc[
    :,
    [
        'release_year', 'duration_ms.x', 'explicit', 
        'mode', 'speechiness', 'acousticness',
        'instrumentalness', 'liveness', 'valence',
        'tempo', 'time_signature'
    ]
] 
# X.head(3)
```

```python
Y = df.loc[
    :,
    [
        'delta: sad to happy',
        'delta: calm to energetic',
        'delta: relaxed to stressed'
    ]
] 
# Y.head()
```

Split into training and not

Creating the model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn import linear_model
from sklearn.multioutput import MultiOutputRegressor

base_estimator = linear_model.LinearRegression()
model_linear = MultiOutputRegressor(base_estimator)
model_linear.fit(X_train, y_train)

model_linear.score(X_train, y_train)
```

Another Model

```python
from sklearn.ensemble import GradientBoostingRegressor

model_gradient_boost = MultiOutputRegressor(GradientBoostingRegressor())
model_gradient_boost.fit(X_train, y_train)
print("Training score")
print(model_gradient_boost.score(X_train, y_train))
print("Test score")
print(model_gradient_boost.score(X_test, y_test))
```

```python
pandas.DataFrame(model_gradient_boost.predict(X_test)).head()
```

```python
y = Y['delta: calm to energetic']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

print("Training score")
print(model_linear.score(X_train, y_train))
print("Test score")
print(model_linear.score(X_test, y_test))
```

```python
y = Y['delta: relaxed to stressed']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()
model_linear.fit(X_train, y_train)
print("Training score")
print(model_linear.score(X_train, y_train))
print("Test score")
print(model_linear.score(X_test, y_test))

```
