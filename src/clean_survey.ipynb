```python
import pandas
```

```python
df_1_survey = pandas.read_csv("../resources/SongSurvey Database - SliderResponses.csv")
```

```python
# df_2_songs = pandas.read_csv("./19332_Spotify_Songs.csv", encoding="MacRoman")
df_2_songs = pandas.read_csv("../out/clean_data.csv")
```

```python
df_merged = pandas.merge(df_1_survey, df_2_songs, how='left', left_on='song_url', right_on='song_link', validate='m:1')
```

```python
df_merged.columns
```

```python

df_merged.head(n=10)
```

```python
emotional_axes = ['sad to happy', 'calm to energetic', 'relaxed to stressed']

for emotional_axis in emotional_axes:
    before = df_merged[f'before: {emotional_axis}']
    after = df_merged[f'after: {emotional_axis}']
    df_merged[f'delta: {emotional_axis}'] = after - before
```

```python
df_merged.to_csv('../out/merged_data.csv')
```
